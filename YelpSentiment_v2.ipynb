```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.neighbors import KNeighborsClassifier
```

```python
def get_score(model, X_train, X_test, y_train, y_test):
    model.fit(X_train, y_train)
    predictions = model.predict(X_test)
    print(classification_report(y_test,predictions)) 
```

```python
get_score(MultinomialNB() , X_train, X_test, y_train, y_test) #Naive Bayes
```

```python
get_score(SVC(), X_train, X_test, y_train, y_test) #Support Vector Machine
```

```python
get_score(LogisticRegression(), X_train, X_test, y_train, y_test) #Logistic Regression
```

```python
get_score(RandomForestClassifier(n_estimators=40), X_train, X_test, y_train, y_test) #RandomForest
```

```python
get_score(KNeighborsClassifier(n_neighbors=5,metric='minkowski') , X_train, X_test, y_train, y_test) #KNeighborsClassifier
```

```python
from sklearn.model_selection import StratifiedKFold
folds = StratifiedKFold(n_splits=3)

scores_logistic = []
scores_svm = []
scores_rf = []

for train_index, test_index in folds.split(X,y):
    X_train, X_test, y_train, y_test = X[train_index], X[test_index], \
                                       y[train_index], y[test_index]
    scores_logistic.append(get_score(LogisticRegression(solver='liblinear',multi_class='ovr'), X_train, X_test, y_train, y_test))  
    scores_svm.append(get_score(SVC(gamma='auto'), X_train, X_test, y_train, y_test))
    scores_rf.append(get_score(RandomForestClassifier(n_estimators=40), X_train, X_test, y_train, y_test))
```
